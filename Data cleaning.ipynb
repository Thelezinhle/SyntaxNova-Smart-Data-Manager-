```python
#importing the necessary library
import pandas as pd
```

```python
#E - Extract: load dataset
data = pd.read_csv(r"C:\Users\Lerat\OneDrive\Documents\App\Online-eCommerce.csv")
```

```python
#T - Tranform: DATA CLEANING:

#delete empty rows
data.dropna(how='all', inplace=True)

# remove extra spaces from column names
data.columns = data.columns.str.strip()

#Convert data types
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors='coerce') # Convert Order_Date to datetime
data['Order_Number'] = pd.to_numeric(data['Order_Number'], errors='coerce')
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')

#remove rows with invalid Order_Date
data = data.dropna(subset=['Order_Date'])

#remove duplicates
data.drop_duplicates(inplace=True)

#creating new column 'Profits'
data['Profits'] = data['Total_Sales'] - data['Total_Cost']
```
